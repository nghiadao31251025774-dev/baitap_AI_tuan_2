```python
2.11#
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

distance = ctrl.Antecedent(np.arange(0, 51, 1), 'distance')
traffic = ctrl.Antecedent(np.arange(0, 101, 1), 'traffic')
demand = ctrl.Antecedent(np.arange(0, 101, 1), 'demand')
weather = ctrl.Antecedent(np.arange(0, 11, 1), 'weather')
rating = ctrl.Antecedent(np.arange(1, 6, 0.1), 'rating')
punctuality = ctrl.Antecedent(np.arange(0, 101, 1), 'punctuality')

price = ctrl.Consequent(np.arange(0, 101, 1), 'price')
bonus = ctrl.Consequent(np.arange(0, 101, 1), 'bonus')

distance['short'] = fuzz.trimf(distance.universe, [0, 0, 3])
distance['medium'] = fuzz.trimf(distance.universe, [2, 5, 8])
distance['long'] = fuzz.trimf(distance.universe, [6, 13, 20])
distance['very_long'] = fuzz.trapmf(distance.universe, [15, 30, 50, 50])

traffic['low'] = fuzz.trimf(traffic.universe, [0, 0, 30])
traffic['medium'] = fuzz.trimf(traffic.universe, [20, 45, 70])
traffic['high'] = fuzz.trimf(traffic.universe, [60, 100, 100])

demand['low'] = fuzz.trimf(demand.universe, [0, 0, 30])
demand['medium'] = fuzz.trimf(demand.universe, [20, 45, 70])
demand['high'] = fuzz.trimf(demand.universe, [60, 100, 100])

weather['good'] = fuzz.trimf(weather.universe, [0, 0, 4])
weather['moderate'] = fuzz.trimf(weather.universe, [3, 5, 7])
weather['bad'] = fuzz.trimf(weather.universe, [6, 10, 10])

rating['poor'] = fuzz.trimf(rating.universe, [1, 1, 2.5])
rating['average'] = fuzz.trimf(rating.universe, [2, 3.25, 4.5])
rating['good'] = fuzz.trimf(rating.universe, [3.5, 5, 5])

punctuality['late'] = fuzz.trimf(punctuality.universe, [0, 0, 50])
punctuality['on_time'] = fuzz.trimf(punctuality.universe, [40, 60, 80])
punctuality['early'] = fuzz.trimf(punctuality.universe, [70, 100, 100])

price['low'] = fuzz.trimf(price.universe, [0, 0, 30])
price['medium'] = fuzz.trimf(price.universe, [25, 50, 75])
price['high'] = fuzz.trimf(price.universe, [60, 80, 100])
price['very_high'] = fuzz.trimf(price.universe, [80, 100, 100])

bonus['none'] = fuzz.trimf(bonus.universe, [0, 0, 10])
bonus['few'] = fuzz.trimf(bonus.universe, [5, 25, 45])
bonus['moderate'] = fuzz.trimf(bonus.universe, [40, 60, 80])
bonus['high'] = fuzz.trimf(bonus.universe, [75, 100, 100])

rules = [
    ctrl.Rule(distance['short'] & traffic['low'] & demand['low'], price['low']),
    ctrl.Rule(distance['short'] & traffic['medium'] & demand['high'], price['medium']),
    ctrl.Rule(distance['medium'] & traffic['high'] & demand['high'], price['high']),
    ctrl.Rule(distance['long'] & traffic['medium'] & weather['good'], price['medium']),
    ctrl.Rule(distance['long'] & traffic['high'] & weather['bad'], price['very_high']),
    ctrl.Rule(distance['very_long'] & traffic['high'] & demand['high'], price['very_high']),
    ctrl.Rule(distance['medium'] & traffic['low'] & demand['low'], price['medium']),
    ctrl.Rule(distance['short'] & traffic['high'] & weather['bad'], price['high']),
    ctrl.Rule(distance['very_long'] & weather['bad'], price['very_high']),
    ctrl.Rule(distance['medium'] & traffic['medium'] & weather['moderate'], price['medium']),
    ctrl.Rule(rating['good'] & punctuality['early'], bonus['high']),
    ctrl.Rule(rating['average'] & punctuality['on_time'], bonus['moderate']),
    ctrl.Rule(rating['poor'] & punctuality['late'], bonus['none']),
    ctrl.Rule(distance['long'] & traffic['high'] & punctuality['on_time'], bonus['high']),
    ctrl.Rule(distance['medium'] & traffic['medium'] & rating['good'], bonus['moderate']),
    ctrl.Rule(rating['poor'] & punctuality['late'], bonus['none']), # Lặp lại luật 13
    ctrl.Rule(distance['very_long'] & weather['bad'] & rating['good'], bonus['high']),
    ctrl.Rule(distance['short'] & rating['average'] & punctuality['on_time'], bonus['few']),
    ctrl.Rule(distance['long'] & traffic['high'] & punctuality['late'], bonus['few']),
    ctrl.Rule(distance['medium'] & weather['moderate'] & rating['good'], bonus['moderate'])
]

grab_system = ctrl.ControlSystem(rules)
grab_sim = ctrl.ControlSystemSimulation(grab_system)

grab_sim.input['distance'] = 20
grab_sim.input['traffic'] = 85
grab_sim.input['demand'] = 90
grab_sim.input['weather'] = 9
grab_sim.input['rating'] = 4.8
grab_sim.input['punctuality'] = 95

grab_sim.compute()

print(f"KẾT QUẢ MÔ PHỎNG GRABBIKE")
print(f"Mức giá dự tính (0-100): {grab_sim.output['price']:.2f}")
print(f"Điểm thưởng dự tính (0-100): {grab_sim.output['bonus']:.2f}")

price.view(sim=grab_sim)
bonus.view(sim=grab_sim)
```

```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl

nhu_cau = ctrl.Antecedent(np.arange(0, 11, 1), 'nhu_cau')
doi_thu = ctrl.Antecedent(np.arange(0, 11, 1), 'doi_thu')
loi_nhuan = ctrl.Antecedent(np.arange(0, 11, 1), 'loi_nhuan')
mua_vu = ctrl.Antecedent(np.arange(0, 11, 1), 'mua_vu')
giam_gia = ctrl.Consequent(np.arange(0, 71, 1), 'giam_gia')
for var in [nhu_cau, doi_thu, loi_nhuan, mua_vu]:
    var.automf(3, names=['Thấp', 'Trung bình', 'Cao'])
giam_gia['Rat_thap'] = fuzzy.trimf(giam_gia.universe, [0, 0, 5])
giam_gia['Thap'] = fuzzy.trimf(giam_gia.universe, [5, 7, 10])
giam_gia['Trung_binh'] = fuzzy.trimf(giam_gia.universe, [10, 15, 20])
giam_gia['Cao'] = fuzzy.trimf(giam_gia.universe, [20, 30, 40])
giam_gia['Rat_cao'] = fuzzy.trimf(giam_gia.universe, [40, 55, 70])
rule1 = ctrl.Rule(nhu_cau['Cao'] & doi_thu['Thấp'] & loi_nhuan['Thấp'], giam_gia['Rat_thap'])
rule2 = ctrl.Rule(nhu_cau['Thấp'] & doi_thu['Cao'] & loi_nhuan['Cao'], giam_gia['Cao'])
rule4 = ctrl.Rule(doi_thu['Cao'] & mua_vu['Cao'] & loi_nhuan['Cao'], giam_gia['Rat_cao'])
rule_custom = ctrl.Rule(loi_nhuan['Cao'] & doi_thu['Trung bình'] & mua_vu['Cao'], giam_gia['Trung_binh'])
system = ctrl.ControlSystem([rule1, rule2, rule4, rule_custom])
sim = ctrl.ControlSystemSimulation(system)
sim.input['nhu_cau'] = 8
sim.input['doi_thu'] = 5
sim.input['loi_nhuan'] = 9
sim.input['mua_vu'] = 9
sim.compute()

print(f"Mức giảm giá đề xuất cho Đồng hồ xa xỉ: {sim.output['giam_gia']:.2f}%")
giam_gia.view(sim=sim) # Lệnh này để vẽ đồ thị kết quả trên Colab
```

```python
2.14#
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

mat_do = ctrl.Antecedent(np.arange(0, 11, 1), 'mat_do')
khan_cap = ctrl.Antecedent(np.arange(0, 11, 1), 'khan_cap')
tai_trong = ctrl.Antecedent(np.arange(0, 11, 1), 'tai_trong')
giao_thong = ctrl.Antecedent(np.arange(0, 11, 1), 'giao_thong')
loi_nhuan = ctrl.Antecedent(np.arange(0, 11, 1), 'loi_nhuan')
ket_hop = ctrl.Consequent(np.arange(0, 11, 1), 'ket_hop')
uu_tien = ctrl.Consequent(np.arange(0, 11, 1), 'uu_tien')
names = ['Thấp', 'Trung bình', 'Cao']
for var in [mat_do, khan_cap, tai_trong, giao_thong, loi_nhuan]:
    var.automf(3, names=names)
ket_hop.automf(3, names=['Ít', 'Một số', 'Nhiều'])
uu_tien.automf(3, names=['Thấp', 'Trung bình', 'Cao'])
rules = [
    ctrl.Rule(mat_do['Cao'] & tai_trong['Thấp'] & giao_thong['Thấp'], ket_hop['Nhiều']),
    ctrl.Rule(mat_do['Trung bình'] & giao_thong['Cao'] & khan_cap['Trung bình'], ket_hop['Một số']),
    ctrl.Rule(tai_trong['Cao'] & mat_do['Cao'] & loi_nhuan['Trung bình'], ket_hop['Một số']),
    ctrl.Rule(mat_do['Thấp'] & khan_cap['Cao'] & giao_thong['Trung bình'], ket_hop['Một số']),
    ctrl.Rule(loi_nhuan['Cao'] & khan_cap['Cao'] & giao_thong['Cao'], ket_hop['Một số']),
    ctrl.Rule(khan_cap['Cao'] & loi_nhuan['Cao'], uu_tien['Cao']),
    ctrl.Rule(khan_cap['Trung bình'] & giao_thong['Trung bình'], uu_tien['Trung bình']),
    ctrl.Rule(khan_cap['Thấp'] & mat_do['Cao'] & loi_nhuan['Thấp'], uu_tien['Thấp']),
    # Thêm luật bổ trợ để tránh lỗi KeyError khi input rơi vào vùng trống
    ctrl.Rule(mat_do['Cao'] & tai_trong['Thấp'], ket_hop['Nhiều'])
]
delivery_ctrl = ctrl.ControlSystem(rules)
delivery_sim = ctrl.ControlSystemSimulation(delivery_ctrl)
delivery_sim.input['mat_do'] = 9
delivery_sim.input['khan_cap'] = 5
delivery_sim.input['tai_trong'] = 2
delivery_sim.input['giao_thong'] = 5
delivery_sim.input['loi_nhuan'] = 5
delivery_sim.compute()
print("======= KẾT QUẢ TỐI ƯU =======")
print(f"Số lượng đơn kết hợp: {delivery_sim.output['ket_hop']:.2f}")
print(f"Mức độ ưu tiên: {delivery_sim.output['uu_tien']:.2f}")
ket_hop.view(sim=delivery_sim)
plt.show()
```
